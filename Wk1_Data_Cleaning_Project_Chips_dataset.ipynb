```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
url='//content/chip_dataset.csv'
df=pd.read_csv(url)
df
```

```python
df=df.drop(['FP16 GFLOPS','FP32 GFLOPS','FP64 GFLOPS'],axis=1)

df.info()
```

```python
df['Release Date'] = pd.to_datetime(df['Release Date'],format="%M/%d/%Y")

df['Release Date']
```

```python
df = df.rename(columns={'Release Date': 'release_date','Process Size (nm)': 'Process_Size_nm',
                        'TDP (W)': 'TDP_W','Die Size (mm^2)': 'Die_Size_mm2',
                        'Transistors (million)': 'Transistors_million','Freq (GHz)': 'Freq_GHz'})

df.info()
```

```python
df.info()
```

```python
df['TDP_W'] = np.floor(pd.to_numeric(df['TDP_W'], errors='coerce'))
df['TDP_W'] = df['TDP_W'].astype('float')

df['Process_Size_nm'] = np.floor(pd.to_numeric(df['Process_Size_nm'], errors='coerce'))
df['Process_Size_nm'] = df['Process_Size_nm'].astype('float')

df['Die_Size_mm2'] = np.floor(pd.to_numeric(df['Die_Size_mm2'], errors='coerce'))
df['Die_Size_mm2'] = df['Die_Size_mm2'].astype('float')

df['Transistors_million'] = np.floor(pd.to_numeric(df['Transistors_million'], errors='coerce'))
df['Transistors_million'] = df['Transistors_million'].astype('float')


df.info()
```

```python
df.Process_Size_nm = df.Process_Size_nm.fillna(df.Process_Size_nm.dropna().mean())
df['Die_Size_mm2']  = df['Die_Size_mm2'].fillna(df['Die_Size_mm2'].dropna().mean())
df.Transistors_million = df.Transistors_million.fillna(df.Transistors_million.dropna().mean())
df.Freq_GHz = df.Freq_GHz.fillna(df.Freq_GHz.dropna().mean())
df.info()
```

```python
df['release_year'] =df['release_date'].dt.year

df[['release_year']]
```

```python



df['transistors_log_10e6'] = np.log10(df.Transistors_million)
df['transistors_log_10e6']
```

```python
plt.figure(figsize=(12,5))



sns.scatterplot(x="release_year",
                    y="transistors_log",
                    data=df,hue='Vendor')
plt.title("Transistors count(Log10) vs Release year")
plt.show()
```

```python
plt.figure(figsize=(12,5))



sns.scatterplot(x="release_year",
                    y="Process_Size_nm",
                    data=df,hue='Vendor')
plt.title("Process Size (nm) vs Release year")
plt.show()
```

```python
plt.figure(figsize=(12,5))



sns.scatterplot(x="Die_Size_mm2",
                    y="Process_Size_nm",
                    data=df,hue='Vendor')
plt.title("Process Size (nm) vs Die Size(mm^2)")
plt.show()
```

```python
df_gc=df[['release_year','Type','Freq_GHz']]
df_gc=df_gc.groupby(['release_year','Type'])[['Freq_GHz']].mean()
df_gc.head()
```

```python
plt.figure(figsize=(12,5))



sns.scatterplot(x="release_year",
                    y="Freq_GHz",
                    data=df_gc,hue='Type')
plt.title("CPU & GPU vs Frequency")
plt.show()
```

```python
df_gp=df[['release_year','Type','transistors_log','Die_Size_mm2','Freq_GHz']]
df_gp=df_gp[df_gp['Type'].isin(['GPU'])]
df_gp=df_gp[['release_year','transistors_log','Die_Size_mm2','Freq_GHz']]
df_gp=df_gp.groupby('release_year')[['transistors_log','Die_Size_mm2','Freq_GHz']].mean().reset_index()
df_gp.head(6)

```

```python
x=df_gp["release_year"]
y=df_gp["transistors_log"]
y1=df_gp["Die_Size_mm2"]
y2=df_gp["Freq_GHz"]

fig,ax=plt.subplots(nrows=1,ncols=3,figsize=(12,5))
ax[0].plot(x,y)
ax[1].plot(x,y1)
ax[2].plot(x,y2)

plt.title("GPU components")
plt.show()
```
